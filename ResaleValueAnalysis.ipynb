```python
import pandas as pd
data = pd.read_csv("CarSales.csv")
```

# Data Cleaning & Feature Engineering

```python
data.dropna(inplace=True)
```

```python
X = data.loc[:,['Brand', 'Body', 'Mileage', 'EngineV', 'Engine Type', 'Registration', 'Year', 'Model']]
y = data.Price
```

```python
X
```

### Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
X.loc[:,['Brand', 'Body', 'Engine Type', 'Registration', 'Model']] = X.loc[:,['Brand', 'Body', 'Engine Type', 'Registration', 'Model']].apply(encoder.fit_transform)
```

```python
X.head()
```
